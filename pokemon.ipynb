Contenu
* nom : Le nom anglais du Pokémon
* japanese_name : Le nom japonais d'origine du Pokémon
* pokedex_number : Le numéro d'entrée du Pokémon dans le National Pokedex
* pourcentage_male : pourcentage d'espèces mâles. Vide si le Pokémon est asexué.
* type1 : Le type principal du Pokémon
* type2 : Le type secondaire du Pokémon
* classification : La classification des Pokémon telle que décrite par le Soleil et la Lune Pokedex
* height_m : Hauteur du Pokémon en mètres
* weight_kg : Le poids du Pokémon en kilogrammes
* capture_rate : Taux de capture du Pokémon 
* étapes de base de l' œuf : le nombre d'étapes nécessaires pour faire éclore un œuf du Pokémon
* capacités : une liste de capacités que le Pokémon est capable d'avoir
* experience_growth : La croissance de l'expérience du Pokémon
* base_happiness : Bonheur de base du Pokémon
against_? : dix-huit caractéristiques qui indiquent la quantité de dégâts subis contre une attaque d'un type particulier
* HP : les HP de base du Pokémon
* attaque : L'attaque de base du Pokémon
* défense : La défense de base du Pokémon
* sp_attack : L'attaque spéciale de base du Pokémon
* sp_defense : La Défense Spéciale de Base du Pokémon
* vitesse : La vitesse de base du Pokémon
* génération : La génération numérotée à laquelle le Pokémon a été introduit pour la première fois
* is_legendary : Indique si le Pokémon est légendaire.
